#Library yang digunakan adalah cv2, numpy, dan matplotlib

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.morphology import thin, skeletonize
from google.colab.patches import cv2_imshow
```

Pada bagian ini kita melakukan read image dengan menggunakan cv2.imread kemudian mengecilkan image dengan resize. pada matriks image, 0 berwarna hitam sedangkan 255 berwarna putih


```
image1 = cv2.imread('rose.PNG')

#image 1
panjang = len(image1)
lebar = len(image1[0])

#print image 1
print('Panjang Horizontal = ',lebar)
print('Panjang Vertikal = ',panjang)

#resize gambar
imSize = np.array(cv2.resize(image1,(200,300),interpolation=cv2.INTER_AREA))

#mencetak gambar
cv2_imshow(image1)
print(image1)
cv2_imshow(imSize)
print(imSize)
```

#gambar gray dicari dengan menggunakan cvtcolor kemudian histogram dari gambargray dicari dengan menggunakan calHist

```
panjang1 = len(imSize)
lebar1 = len(imSize[0])
print('Panjang Horizontal = ',lebar1)
print('Panjang Vertikal = ',panjang1)


gambargray = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
cv2_imshow(gambargray)
print(gambargray)

#print('Panjang Horizontal 2 = ',lebar1)
#print('Panjang Vertikal 2 = ',panjang1)
#print(panjang, 'px x',lebar 'px')
grayhist = cv2.calcHist([gambargray],[0],None,[255],[0,255])
plt.plot(grayhist,color='r')
plt.show()
```

#yang kurang dari sama dengan 200 akan dianggap hitam, dan sisanya diatas 200 dianggap putih

```
gambarbiner = gambargray
#traditional thresholding
for i in range(panjang):
    for j in range(lebar):
        if(gambarbiner[i,j] <= 200):
            gambarbiner[i,j] = 0;
        else:
            gambarbiner[i,j] = 255;

plt.imshow(cv2.cvtColor(gambarbiner, cv2.COLOR_BGR2RGB))
```

#disini kita akan melakukan deteksi tepi, untuk melakukan deteksi tepi digunakan library dari edge detection yaitu canny

```
edged = cv2.Canny(gambarbiner, 30, 200) #Perform Edge detection
plt.figure(1)
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
```
